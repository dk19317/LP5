```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv('Boston.csv')
df.head(10)
```

## Drop unwanted columns with NaN

```python
df.drop(columns=['Unnamed: 15','Unnamed: 16'],inplace=True)
```

```python
df.drop(columns=['CAT. MEDV'],inplace=True)
```

## Checking correlation with target variable MEDV

```python
df.corr()['MEDV'].sort_values()
```

```python
X = df.loc[:,['LSTAT','PTRATIO','RM']]
Y = df.loc[:,"MEDV"]
X.shape,Y.shape
```

## Preparing training and testing data set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=10)
```

## Normalizing training and testing dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

## Preparing model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python

model = Sequential()
```

```python
model.add(Dense(128,input_shape=(3,),activation='relu',name='input'))
model.add(Dense(64,activation='relu',name='layer_1'))
model.add(Dense(1,activation='linear',name='output'))
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
model.fit(x_train,y_train,epochs=100,validation_split=0.05)
```

```python
output = model.evaluate(x_test,y_test)
```

```python
print(f"Mean Squared Error: {output[0]}"
      ,f"Mean Absolute Error: {output[1]}",sep="\n")
```

```python
y_pred = model.predict(x=x_test)
```

```python
print(*zip(y_pred,y_test))
```

##

```python
import pandas as pd

# Suppose you have y_pred and y_test as lists or arrays
data = list(zip(y_pred, y_test))

# Create a DataFrame
df = pd.DataFrame(data, columns=["Predicted", "Actual"])

# Display nicely
print(df.to_string(index=False))
```

## Scatter Plot Predicted vs Actual:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df["Actual"], df["Predicted"], color='blue', label="Predicted vs Actual")
plt.plot([df["Actual"].min(), df["Actual"].max()], [df["Actual"].min(), df["Actual"].max()], 'r--', label="Perfect Prediction Line")
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs Actual Scatter Plot')
plt.legend()
plt.grid(True)
plt.show()
```

## Calculate Error Metrics:

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(df["Actual"], df["Predicted"])
mae = mean_absolute_error(df["Actual"], df["Predicted"])

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```
